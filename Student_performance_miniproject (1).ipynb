```python
#2 combine the two dataframe Horizantal and store the variable df1 (use concat).
df1=pd.concat([stu_info,stu_marks],axis=1) #concating two dataframes horizontally
df1
```

```python
#3.Combine the two dataframe vertically column and store the df2(use concat).
df2=pd.concat([stu_info,stu_marks],axis=0)  #concating two dataframes vertically
df2
```

```python
#4.Combine the two dataframes based on all join types (inner, outer, left, right) using a common column from both dataframes and store the r

d1=pd.merge(stu_info,stu_marks,how='inner',left_index=True,right_index=True) #merging two dataframes using inner join
d1
```

```python
d2=pd.merge(stu_info,stu_marks,how='outer',right_index=True,left_index=True) #merging two data frames using outer join
d2
```

```python
d3=pd.merge(stu_info,stu_marks,how='left',right_index=True,left_index=True) #merging two dataframes using left join
d3
```

```python
d4=pd.merge(stu_info,stu_marks,how='right',right_index=True,left_index=True) #merging two dataframes using right join
d4
```

```python
#6. Find the average marks of all subjects and store them in a column named `'avg_marks'`.
d1['avg_marks'] = (d1['math score']+ d1['reading score'] + d1['writing score'])/3
d1
```

```python
#7. Find the `id` and `gender` corresponding to the maximum value of `avg_marks`.
max_index = d1['avg_marks'].idxmax()
max_index
```

```python
# Retrieve the id and gender for the maximum avg_marks
result = d1.loc[max_index, ['id', 'gender']]

print(result)
```

```python
#8. Find the `id` and `gender` corresponding to the minimum value of `avg_marks`.
min_index = d1['avg_marks'].idxmin()
min_index
```

```python
# Retrieve the id and gender for the maximum avg_marks
result = d1.loc[min_index, ['id', 'gender']]
result
```

```python
#9. Find the sum of `avg_marks` based on race/ethnicity.
race_total=d1.groupby('race/ethnicity')['avg_marks'].sum()
race_total
```

```python
#10. Find the sum, average, maximum , minimum, and count of the math scores based on gender.
stu_math = d1.groupby('gender')['math score'].agg(['sum', 'mean', 'max', 'min', 'count'])
stu_math
```

```python
#11. Find the sum, average, maximum , minimum, and count of the reading scores based on gender.
stu_reading = d1.groupby('gender')['reading score'].agg(['sum', 'mean', 'max', 'min', 'count'])
stu_reading
```

```python
#12. Find the sum, average, maximum , minimum, and count of the writing scores based on gender.
stu_writing = d1.groupby('gender')['writing score'].agg(['sum', 'mean', 'max', 'min', 'count'])
stu_writing
```

```python
#13. Sort the data based on the writing scores
writing_sorted=d1.sort_values(by='writing score')
writing_sorted
```

```python
#14. Find the count of males and females and their respective percentages.
# Count the number of each gender
gender_counts = d1['gender'].value_counts()

# Calculate the total number of entries
total_count = len(d1)

# Calculate the percentage of each gender
gender_percentages = (gender_counts / total_count) * 100

# Combine counts and percentages into a DataFrame for better readability
gender_summary = pd.DataFrame({
    'Count': gender_counts,
    'Percentage': gender_percentages
})

print(gender_summary)
```

```python
#15. Find the count of each race/ethnicity and their respective percentages.
# Count the number of each gender
race_eth_counts = d1['race/ethnicity'].value_counts()

# Calculate the total number of entries
total_count = len(d1)

# Calculate the percentage of each gender
race_eth_percentages = (race_eth_counts / total_count) * 100

# Combine counts and percentages into a DataFrame for better readability
race_eth_summary = pd.DataFrame({
    'Count': race_eth_counts,
    'Percentage': race_eth_percentages
})

print(race_eth_summary)
```
